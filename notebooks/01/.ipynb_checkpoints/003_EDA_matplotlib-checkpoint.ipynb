```python
import pandas as pd
import matplotlib.pyplot as plt
from dateutil.parser import parse
```

```python
f = "../../data/Irradiancia.csv"
pd.read_csv(f,index_col=0,parse_dates=True)
```

```python
f = "../../data/Irradiancia.parquet"
I = pd.read_parquet(f)
#Arreglo los datos
I.sort_index(inplace=True)
I = I[I.index.year>2022]
```

```python
I.plot(subplots=True, figsize=(12,8))
```

```python
fig, ax = plt.subplots(figsize=(12,4))

f1 = parse("2024-03-10")
f2 = f1 + pd.Timedelta("1D")

ax.plot(I.I_glo_Avg,label="I_glo")
ax.plot(I.I_dir_Avg,label="I_dir")
ax.plot(I.I_dif_Avg,label="I_dif")

ax.set_xlim(f1,f2)
ax.set_ylim(0,1200)
ax.legend()
```
