```python
import pandas as pd
```

```python
f = "../../data/estacion_meteorologica.zip"
# f = "../../data/prueba.txt"
estacion = pd.read_csv(f,skiprows=[0,2,3],index_col=0,parse_dates=True)
estacion.I_glo_Avg = estacion.I_glo_Avg.astype(float)
estacion.I_dir_Avg = estacion.I_dir_Avg.astype(float)
del estacion["RECORD"]
estacion.info()
```

```python
columnas = estacion.columns
Is = [columna for columna in columnas if ("Avg" and "I_") in columna]

Is
irradiancia = estacion[Is]
irradiancia
```

```python
irradiancia.to_csv("../../data/Irradiancia.csv")
irradiancia.to_parquet("../../data/Irradiancia.parquet")
```
